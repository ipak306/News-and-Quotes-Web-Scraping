```python
#สร้าง list total_url เพื่อเก็บข้อมูลของ url หน้า web page ทั้งหมดที่ต้องการดึงข้อมูล 
total_url = []
i = 1
while True: #วนลูปหา url ของหน้า web page ทั้งหมด ใช้วิธีการร้องขอด้วย requests 
    url = "https://quotes.toscrape.com/page/{}/".format(i)
    r = requests.get(url)
    soupp = soup(r.content,'html.parser')
    test = soupp.findAll('div',{'class':'col-md-8'})
    #ถ้าหากหน้า url มีเนื้อหาว่า No quotes found! ให้ break
    if test[1].text == '\n\nNo quotes found!\n\n    \n\n\n← Previous\n\n\n\n': 
        break
    else:
        total_url.append(url)
        i += 1
        
```

โดยจะทำการสร้าง function เพื่อดึงข้อมูลที่ต้องการภายในหน้า web page ของแต่ละ url 



#### ฟังก์ชั่น scrape_quote(data) ดึงข้อมูลส่วนของเนื้อหา quote


```python
def scrape_quote(data):
    quote = []
    desc = data.findAll('span',{'class':'text'}) #หา tag html ที่มี class = text ที่มี เนื้อหา quote อยู่
    # วนลูปเพื่อเก็บ quote ไว้ใน list quote
    for a in range(0,len(desc)):
        quote.append(desc[a].text.replace('\'','').replace('“','').replace('”',''))
        
    return quote
```

#### ฟังก์ชั่น scrape_tags(data) ดึงข้อมูลส่วนของ tags ที่ถูกติดให้ของแต่ละ quotes 


```python
def scrape_tags(data):
    tags = []
    test3 = data.findAll('meta',content=True)  #หา tag html meta ที่มีส่วนของ tags อยู่
    # วนลูปเพื่อเก็บ tags ไว้ใน list tags
    for b in range(0,len(test3)):
         tags.append(test3[b]['content'])
            
    return tags
```

#### ฟังก์ชั่น scrape_name_author(data) ดึงข้อมูลในส่วนของชื่อผู้เขียน


```python
def scrape_name_author(data):
    name_author = []
    test2 = data.findAll('small',{'class':'author'}) #หา tag html small ที่มี class = author มีเนื้อหาส่วนของ ชื่อผู้เขียน อยู่
    # วนลูปเพื่อเก็บ ชื่อผู้เขียน ไว้ใน list name_author
    for j in range(0,len(test2)):
        name_author.append(test2[j].text)
        
    return name_author
```

#### ฟังก์ชั่น scrape_hbd_author(data2) ดึงข้อมูลในส่วนของวันเกิดของผู้เขียน


```python
def scrape_hbd_author(data2):
    hbd_author = []
    #หา tag html span ที่มี class = author-born-date มีเนื้อหาส่วนของ วันเกิดผู้เขียน อยู่
    hbd_au = data2.findAll('span',{'class': 'author-born-date'})[0].text 
    # วนลูปเพื่อเก็บ วันเกิดผู้เขียนไว้ใน list hbd_author
    hbd_author.append(hbd_au)

    return hbd_author
```

#### ฟังก์ชั่น scrape_hbdlo_author(data2) ดึงข้อมูลในส่วนของสถานที่เกิดของผู้เขียน


```python
def scrape_hbdlo_author(data2):
    hbdlo_author = []

    #หา tag html span ที่มี class = author-born-location มีเนื้อหาส่วนของ สถานที่เกิดผู้เขียน อยู่
    hbdlo_au = data2.findAll('span',{'class': 'author-born-location'})[0].text
    hbdlo_author.append(hbdlo_au[3:])
    # วนลูปเพื่อเก็บ สถานที่เกิดผู้เขียนไว้ใน list hbdlo_author
    return hbdlo_author
```

#### ฟังก์ชั่น scrape_des_author(data2) ดึงข้อมูลในส่วนของประวัติของผู้เขียน


```python
def scrape_des_author(data2):
    des_author = []
    #หา tag html div ที่มี class = author-description มีเนื้อหาส่วนของ ประวัติผู้เขียน อยู่
    des_au = data2.findAll('div',{'class': 'author-description'})[0].text
    # วนลูปเพื่อเก็บ ประวัติผู้เขียนไว้ใน list des_author
    des_author.append(des_au.replace('\n        ','').replace('    \n    ',''))
    return des_author
```

#### ฟังก์ชั่น scrape_link_about(data) ดึงข้อมูล url link about ของแต่ละ quote




ซึ่งภายใน link about จะมีหน้า web ดังนี้


โดยจะทำการเรียกใช้ฟังก์ชั่น 
scrape_hbd_author(data2),  scrape_hbdlo_author(data2), scrape_des_author(data2) เพื่อดึงข้อมูลภายในหน้า web page link about นี้

```python
def scrape_link_about(data):   
    check_link_about = []
    check = []
    
    #ให้ test4 ค้นหา tag html ที่เป็น a
    test4 = data.findAll('a')
    
    #วนลูปเพื่อหา url link about ของแต่ละ quote โดยข้อมูลที่ได้จะได้เพียง "/author/Albert-Einstein" แล้วเก็บข้อมูลไว้ใน list check_link_about
    for k in range(0,len(test4)):
        if test4[k].text == '(about)':
            check_link_about.append(test4[k]['href'])
        else:
            pass
    #วนลูปเพื่อสร้าง url ที่สมบูรณ์ โดย m คือข้อมูลภายใน list check_link_about เอามาต่อเข้ากับ 'https://quotes.toscrape.com' เพื่อได้เป็น url link about
    #ของแต่ละ quote ในการนำไปดึงข้อมูลภายใน url นั้นได้ แล้วเก็บไว้ใน list check
    for m in check_link_about:
        total_link_au = 'https://quotes.toscrape.com'+ m
        check.append(total_link_au)
     # วนลูปเพื่อ Parsing html ของแต่ละ url ใน list check
    for n in check:
        url = n + '/'
        response = urllib.request.urlopen(url)
        html = response.read().decode()
        data2 = soup(html,'html.parser')
        
        #เรียกใช้ฟังก์ชั่น scrape_hbd_author(data2) ดึงข้อมูลวันเกิดของผู้เขียน
        for i in scrape_hbd_author(data2):
            all_hbd_author.append(i)
            
        #เรียกใช้ฟังก์ชั่น scrape_hbdlo_author(data2) ดึงข้อมูลสถานที่เกิดของผู้เขียน
        for j in scrape_hbdlo_author(data2):
            all_hbdlo_author.append(j)           

        #เรียกใช้ฟังก์ชั่น scrape_des_author(data2) ดึงข้อมูลประวัติของผู้เขียน
        for f in scrape_des_author(data2):
            all_des_author.append(f)
```

พอสร้างังก์ชั่นดึงข้อมูลจากหน้า web page link about ได้เสร็จแล้วจะทำการดึงข้อมูลจากหน้าเว็บหลักของแต่ละ quote 

```python
all_quote = [] #เก็บ quote ทั้งหมด
all_tags = [] #เก็บ tags ทั้งหมด
all_name_author = [] #เก็บชื่อผู้เขียนทั้งหมด
all_hbd_author = [] #เก็บวันเกิดผู้เขียนทั้งหมด
all_hbdlo_author = [] #เก็บสถานที่เกิดผู้เขียนทั้งหมด
all_des_author = [] #เก็บประวัติผู้เขียนทั้งหมด

#วนลูป url ของหน้าเว็บหลักทั้ง 10 url
for i in total_url:
    #parsing html ของแต่ละ url
    response = urllib.request.urlopen(i)
    html = response.read().decode()
    response.close()
    data = soup(html,'html.parser')
    # วนลูปในฟังก์ชั่น scrape_quote(data) เพื่อเก็บข้อมูลเนื้อหา quote
    for i in scrape_quote(data):
        all_quote.append(i)
    # วนลูปในฟังก์ชั่น scrape_tags(data) เพื่อเก็บข้อมูล tags
    for b in scrape_tags(data):
        all_tags.append(b)
    # วนลูปในฟังก์ชั่น scrape_name_author(data) เพื่อเก็บข้อมูล ชื่อผู้เขียน
    for j in scrape_name_author(data):
        all_name_author.append(j)
    # เรียกใช้ฟังก์ชั่น scrape_link_about(data) เพื่อดึงข้อมูลภายในหน้า web page ของ link about
    # ซึ่งจะได้ข้อมูล วันเกิดผู้เขียน สถานที่เกิดผู้เขียน และ ประวัติผู้เขียน
    scrape_link_about(data) 
```

```python
#ทำการแปลงจาก list to dataframe
all_quote = pd.Series(all_quote)
all_tags = pd.Series(all_tags)
all_name_author = pd.Series(all_name_author)
all_hbd_author = pd.Series(all_hbd_author)
all_hbdlo_author = pd.Series(all_hbdlo_author)
all_des_author = pd.Series(all_des_author)
# รวม dataframe ทุกอันเข้าด้วยกัน
df = pd.concat([all_quote, all_tags,all_name_author,all_hbd_author,all_hbdlo_author,all_des_author], axis=1)

#เปลี่ยนชื่อของแต่ละ columns ภายใน dataframe
df = df.rename(columns={0 : 'quote', 1 : 'tags', 2 : 'name_author', 3 : 'hbd_author', 4 : 'hbdlo_author', 5 : 'des_author'})

# check ว่าแต่ละคอลัมน์มีค่า null ไหม
df.isnull().sum()
```

## Text Pre- Process

ทำ Text Pre- Process ของข้อมูล feature ซึ่งเป็น unstructure data(text) ให้อยู่ในรูปของ structure data(dataframe ตารางที่เข้าใจได้ง่ายนำไปใช้วิเคราะห์ข้อมูลต่อไปได้) ด้วยกระบวนการดังนี้
1. tokenisation เป็นกระบวนการตัดคำจากประโยคภาษาอังกฤษจาก white space ให้จากประโยค กลายเป็น คำ
2. case conversation เป็นกระบวนการแปลงจากตัวอักษรพิมพ์ใหญ่ให้เป็นพิมพ์เล็ก
3. filter stop words กระบวนการกำจัดคำที่ไม่มีความหมาย ยกตัวอย่างเช่น a and the
4. Lemmatise และ POS tag กระบวนการลดรูปคำ กลายเป็น common form ของคำ โดยจะใช้ POS tag เข้ามาใช้ POS tag จะเป็นการหาว่าคำมีหน้าที่อย่างไรในประโยคภาษาอังกฤษ เป็นกริยา หรือ คำนามของประโยคให้ลดรูปคำได้เก่งขึ้น
5. TF-IDF weight การถ่วงน้ำหนักของคำเพื่อหาว่าคำนั้นมีพลังในการแยกแยะเอกสารมากน้อยขนาดไหนจึงใช้ weight เข้ามาช่วยถ่วงน้ำหนักคำว่ามีความสำคัญมากน้อยขนาดไหน

```python
#Tokenizer and WordNet Lemmatizer with POS
#สร้าง function lemma_tokenizer_w_pos_tag(text) ในการทำ text pre process
def lemma_tokenizer_w_pos_tag(text):
    
    # สร้างฟังก์ชั่นในการแปลง POS tag ต่างๆ สำหรับใช้กับ Lemmatizer 
    def convert_tags(tag):
        if tag == 'vbd' or tag == 'vbg' or tag == 'vbz': #ถ้า tag = vbd,vbg,vbz ให้ return เป็น v(verb)
            return 'v'
        else:  #ถ้าไม่ใช่ให้ return เป็น n(noun)
            return 'n'

    #ทำการตัดคำด้วย standard_tokenizer
    standard_tokenizer = CountVectorizer().build_tokenizer()
    #print(standard_tokenizer)
    tokens = standard_tokenizer(text)
    #print(tokens)
    
    #ทำการตัดคำด้วย pos tag 
    tokens_with_pos_tag = nltk.pos_tag(tokens)
    #print(tokens_with_pos_tag)
    
    # ใช้ NLTK ในการ lemmatisation(ลดรูปคำแต่ละคำ)
    lemmatizer = nltk.WordNetLemmatizer()
    lemma_tokens = []
    
    for token in tokens_with_pos_tag: #วนลูปของแต่ละ token
    #print(token)
    #print(token[1].lower())
        new_tag = convert_tags(token[1].lower())  #เรียกใช้ฟังก์ชั่น convert_tags ในการแปลง POS tag และทำให้เป็นตัวอักษรพิมพ์เล็กทั้งหมด
    #print(new_tag)
    #dd = lemmatizer.lemmatize(token[0]) 
    #print(dd)
        lemma_tokens.append(lemmatizer.lemmatize(token[0], new_tag)) #ทำการลดรูปคำแล้วเก็บไว้ใน lemma_tokens
    
    #print(lemma_tokens)
    return lemma_tokens
```

```python
# ใช้ CountVectorizer ในการทำ Bag of Word ของ feature
# โดยการให้ stop_words = english คือเอาคำที่อยู่ใน list english ออก
# min_df = 0 คือ ถ้าคำนั้นปรากฎในเอกสารน้อยกว่า 0 ครั้งจะถูก remove 
# tokenizer = lemma_tokenizer_w_pos_tag คือใช้การตัดคำด้วยการเรียกใช้ ฟังก์ชั่น lemma_tokenizer_w_pos_tag
vectorizer = CountVectorizer(stop_words="english",min_df = 0,tokenizer=lemma_tokenizer_w_pos_tag)
X = vectorizer.fit_transform(feature) 
print(X.shape)
```

### train test 90:10
การแบ่ง train test ควรพิจารณาจากจุดประสงค์ต่างๆ รวมถึงปริมาณของข้อมูลที่มี เนื่องจากข้อมูลมีจำนวนน้อยเพียง 100 rows จึงเลือกการแบ่งข้อมูลด้วย traintest 90:10  หลักการของการแบ่งแบบ 90:10 คือ ให้ train data มีจำนวน 90 examples และ test data 10 examples เพื่อให้ model ได้ทำการเรียนรู้จากข้อมูล train ที่มีจำนวนมากกว่าเพื่อนำไปใช้สร้างโมเดลได้อย่างมีประสิทธิภาพ และแบ่งข้อมูล test ไว้เพียง 10 examples เพื่อใช้สำหรับการวัดผลของโมเดล

```python
# สร้าง function เพื่อใช้ในการวิเคราะห์ และทำนายแต่ละโมเดล
def create_model(model, parameters,feature):
    total2 = [] # สร้าง list ไว้เก็บค่า accuracy precision และ recall ของทุก tags
    
    for i in label_top15: # วนลูปใน tags 15 ตัวแรกที่มากที่สุด
        X_train , X_test , y_train , y_test = train_test_split(feature, label_top15[i], test_size = 0.1, random_state = 0)
        # แบ่งชุดข้อมูลออกเป็น 2 ชุด trian = 90 test = 10 
        
        total = []
        
        #ทำการหา best parameter ที่ดีที่สุดด้วย GridsearchCV 
        skf = StratifiedKFold(n_splits=5, shuffle = True, random_state = 1001)
        # ใส่ค่า model ที่ต้องการใช้ และ parameters ของ model ที่ต้องการ
        tune = GridSearchCV(model, parameters, verbose = 3, cv = skf.split(X_train,y_train), scoring='accuracy', n_jobs = -1)
        tune.fit(X_train, y_train)

        bestparams = tune.best_params_
        model.set_params(**bestparams) # set ค่า best parameter ที่ได้จาก GridSearchCV กับ model

        model.fit(X_train, y_train) #สร้าง modelด้วย X_train, y_train
        predicted = model.predict(X_test) # ให้ model ทำนายด้วย X_test

        accuracy = accuracy_score(y_test, predicted) # วัดประสิทธิภาพโมเดลของข้อมูล y_test กับ predicted ด้วย accuracy

        precision = precision_score(y_test,predicted, average='binary') # วัดประสิทธิภาพโมเดลด้วย precision #binary คือแสดงค่า precision ฝั่งบวก

        recall = recall_score(y_test,predicted, average='binary') # วัดประสิทธิภาพโมเดลด้วย recall #binary คือแสดงค่า precision ฝั่งบวก
        
        #เพิ่มข้อมูลของ accuracy precision recall ของ tags นั้นใน list total
        total.append('%.2f'%accuracy) 
        total.append('%.2f'%precision)
        total.append('%.2f'%recall)
        total2.append(total) #เพิ่มข้อมูล accuracy precision recall ของแต่ละ tags ใน list total2
    
    # นำ list total2 มาสร้างเป็น dataframe สะดวกต่อการอ่านผล
    df = pd.DataFrame(total2, index=[i for i in label_top15],columns = ['accuracy','precision','recall']) 
    
    return  df
```

จะทำด้วยวิธี grid search CV ซึ่งเป็นวิธีในการหาค่าที่ดีที่สุดของพารามิเตอร์ 
หลักการคือ เริ่มจากกำหนดค่าของแต่ละพารามิเตอร์ที่ต้องการจะหา แล้วรันโมเดล และจะต้องทำแบบนี้จนกว่าจะหมดพารามิเตอร์ที่ต้องการหา โดยต้องเปลี่ยนค่าพารามิเตอร์ไปเรื่อยๆจนได้พารามิเตอร์ที่ทำให้โมเดลมีค่า accuracy สูงที่สุด

```python
# สร้างฟังก์ชั่น parameters KNN ไว้เรียกใช้เมื่อทำการ tune hyperparameter ของ model KNN
def par_KNN():
    parameters = { 
        'n_neighbors' : list(np.arange(3,70,2)),  # กำหนดจำนวนเพื่อนบ้านที่ใกล้ที่สุดตั้งเเต่ 3 - 70 จำนวน เพื่อ return ค่าที่เหมาะสมที่สุดสำหรับโมเดล
        'weights' : ['uniform', 'distance'], # เพื่อเลือกพารามีเตอร์ที่มีผลต่อการคำนวณน้ำหนักในเเต่ละจุดข้อมูล เพื่อลดปัญหาการนำข้อมูลเข้าผิดกลุ่ม เนื่องจากบางกรณี หากเราไม่มีการคำนวณค่าน้ำหนักของข้อมูลอื่นๆ การกำหนดเพียงค่า k อย่างเดียวอาจทำให้้ได้ผลของการเเบ่งกลุ่มที่ผิดพลาดได้
        'metric' : ['euclidean', 'manhattan'] # เพื่อดูว่าต้องใช้การคำนวนระยะทางเเบบใดเพื่อได้ model ที่มีประสิทธิภาพมากที่สุด
        }
    return parameters
```

XGBoost เป็นหลักการของ boosting ใน ensemble method จะมีการ train ข้อมูลหลายๆรอบ โดยหลักการของ boosting จะทำการดูแต่ละรอบว่าในรอบนั้นข้อมูลตัวไหนที่ทำนายผิดจะให้ข้อมูลตัวนั้นถูก train ในรอบต่อไปมากขึ้นซึ่งการทำ boosting ก็เป็นอีกหนึ่งวิธีที่ทำให้ accuracy นั้นดีมากขึ้นจึงเลือกโมเดลนี้เป็นอีกโมเดลหนึ่ง
